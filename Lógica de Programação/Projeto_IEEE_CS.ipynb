## üé≤ Estrutura de Dados - Lista Duplamente Encadeada

````python
class ListaException(Exception):
    def __init__(self, msg) -> None:
        super().__init__(msg)


class Node:
    def __init__(self, carga):
        self.__carga = carga
        self.__prox = None
        self.__ant = None

    @property
    def carga(self):
        return self.__carga

    @property
    def prox(self):
        return self.__prox

    @property
    def ant(self):
        return self.__ant

    @carga.setter
    def carga(self, novaCarga):
        self.__carga = novaCarga

    @prox.setter
    def prox(self, novoProx):
        self.__prox = novoProx

    @ant.setter
    def ant(self, novoAnt):
        self.__ant = novoAnt


class Lista:
    def __init__(self) -> None:
        self.__head = None
        self.__tamanho = 0

    def empty(self) -> bool:
        return self.__tamanho == 0

    def size(self) -> int:
        return self.__tamanho

    def search(self, valor) -> int:
        if self.empty():
            raise ListaException('Lista est√° vazia')

        pointer = self.__head
        contador = 1

        while pointer is not None:
            if pointer.carga == valor:
                return contador
            pointer = pointer.prox
            contador += 1

        raise ListaException('O valor informado n√£o est√° na lista')

    def element(self, posicao) -> int:
        try:
            assert posicao > 0
            if posicao > self.__tamanho:
                raise ListaException('Posi√ß√£o inv√°lida')

            atual = self.__head
            for _ in range(posicao - 1):
                atual = atual.prox

            return atual.carga
        except TypeError:
            raise ListaException('O argumento deve ser do tipo inteiro')
        except AssertionError:
            raise ListaException('Posi√ß√£o negativa n√£o √© v√°lida')
        except:
            raise

    def insert(self, posicao, carga) -> None:
        try:
            assert posicao > 0
            if posicao > self.__tamanho + 1:
                raise ListaException('Posi√ß√£o inv√°lida')

            novoNo = Node(carga)

            if posicao == 1:  # Inserir no in√≠cio
                if self.empty():
                    self.__head = novoNo
                else:
                    novoNo.prox = self.__head
                    self.__head.ant = novoNo
                self.__head = novoNo
            elif posicao == self.__tamanho + 1:  # Inserir no final
                atual = self.__head
                while atual.prox is not None:
                    atual = atual.prox
                atual.prox = novoNo
                novoNo.ant = atual
            else:  # Inserir no meio
                atual = self.__head
                for _ in range(posicao - 2):
                    atual = atual.prox

                novoNo.prox = atual.prox
                novoNo.ant = atual
                atual.prox.ant = novoNo
                atual.prox = novoNo

            self.__tamanho += 1
        except TypeError:
            raise ListaException('O argumento deve ser do tipo inteiro')
        except AssertionError:
            raise ListaException('Posi√ß√£o negativa n√£o √© v√°lida')
        except:
            raise

    def remove(self, posicao) -> int:
        try:
            assert posicao > 0
            if posicao > self.__tamanho:
                raise ListaException('Posi√ß√£o inv√°lida')

            if posicao == 1:  # Remover o primeiro elemento
                valor = self.__head.carga
                if self.__head.prox is None:  # Se s√≥ h√° um elemento na lista
                    self.__head = None
                else:
                    self.__head.prox.ant = None
                    self.__head = self.__head.prox
            elif posicao == self.__tamanho:  # Remover o √∫ltimo elemento
                atual = self.__head
                while atual.prox is not None:
                    atual = atual.prox
                valor = atual.carga
                atual.ant.prox = None
            else:  # Remover um elemento do meio
                atual = self.__head
                for _ in range(posicao - 1):
                    atual = atual.prox

                valor = atual.carga
                atual.ant.prox = atual.prox
                atual.prox.ant = atual.ant

            self.__tamanho -= 1
            return valor
        except TypeError:
            raise ListaException('O argumento deve ser do tipo inteiro')
        except AssertionError:
            raise ListaException('Posi√ß√£o negativa n√£o √© v√°lida')
        except:
            raise

    def modify(self, posicao, novoValor) -> bool:
        try:
            assert posicao > 0
            if posicao > self.__tamanho:
                raise ListaException('Posi√ß√£o inv√°lida')

            atual = self.__head
            for _ in range(posicao - 1):
                atual = atual.prox

            atual.carga = novoValor
            return True
        except TypeError:
            raise ListaException('O argumento deve ser do tipo inteiro')
        except AssertionError:
            raise ListaException('Posi√ß√£o negativa n√£o √© v√°lida')
        except:
            raise

    def __str__(self) -> str:
        result = []
        atual = self.__head
        while atual is not None:
            result.append(str(atual.carga))
            atual = atual.prox
        return '[' + ', '.join(result) + ']'

    def print(self) -> None:
        print(self.__str__())
````

 ## üìú Objetos Necess√°rios na Aplica√ß√£o

````python
#from ListaDuplamenteEncadeada import *
import pandas as pd
import os

class Usuario:
    def __init__(self, cpf:str, nome:str, telefone:str, endereco:str):
            self.__cpf = cpf
            self.__nome = nome
            self.__telefone = telefone
            self.__endereco = endereco

    # Getters
    def getCpf(self) -> str:
        return self.__cpf

    def getNome(self) -> str:
        return self.__nome

    def getTelefone(self) -> str:
        return self.__telefone

    def getEndereco(self) -> str:
        return self.__endereco

    # Setters
    def setCpf(self, novo_cpf):
        self.__cpf = novo_cpf

    def setNome(self, novo_nome):
        self.__nome = novo_nome

    def setTelefone(self, novo_telefone):
        self.__telefone = novo_telefone

    def setEndereco(self, novo_endereco):
        self.__endereco = novo_endereco

    # Print
    def __str__(self) -> str:
        return f"( CPF: {self.__cpf}, Nome: {self.__nome}, Telefone: {self.__telefone}, Endere√ßo: {self.__endereco} ) \n"

    def print(self) -> None:
        print(self.__str__())


class Cadastro:
    def __init__(self):
        self.__dados = Lista()

    # M√©todos
    def adicionarUsuario(self, cpf, nome, telefone, endereco) -> None:
        # Verifica se j√° existe um usu√°rio com o mesmo CPF
        for i in range(1, self.__dados.size() + 1):
            usuario = self.__dados.element(i)
            if usuario.getCpf() == cpf:
                print("J√° existe um usu√°rio com esse CPF.\n")
                return

        self.__dados.insert(self.__dados.size() + 1, Usuario(cpf, nome, telefone, endereco))

    def removerUsuario (self, cpf) -> None:
        for i in range(1, self.__dados.size() + 1):
            usuario = self.__dados.element(i)
            if usuario.getCpf() == cpf:
                self.__dados.remove(i)
                return

        print("Usu√°rio n√£o encontrado.\n")


    def atualizarUsuario(self, cpf, nome, telefone, endereco) -> None:
        for i in range(1, self.__dados.size() + 1):
            usuario = self.__dados.element(i)
            if usuario.getCpf() == cpf:
                usuario.setNome(nome)
                usuario.setTelefone(telefone)
                usuario.setEndereco(endereco)
                return

        print("Usu√°rio n√£o encontrado.\n")


    def consultarUsuario(self, cpf) -> object:
        for i in range(1, self.__dados.size() + 1):
            usuario = self.__dados.element(i)
            if usuario.getCpf() == cpf:
                return usuario

        return None  # Retorna None se o usu√°rio n√£o for encontrado

    def salvarArquivo(self):
        try:
            with open('cadastros.csv', 'w') as arquivo:
                arquivo.write('CPF;Nome;Telefone;Endere√ßo\n')
                for i in range(1, self.__dados.size() + 1):
                    usuario = self.__dados.element(i)
                    arquivo.write(f'{usuario.getCpf()};{usuario.getNome()};{usuario.getTelefone()};{usuario.getEndereco()}\n')
        except Exception as e:
            print(f"Erro ao salvar o arquivo: {e}")

    def lerArquivo(self):
        try:
            with open('cadastros.csv', 'r') as arquivo:
                linhas = arquivo.readlines()
                if len(linhas) == 1:
                    print('N√£o existem registros de cadastro.')
                else:
                    for linha in linhas[1:]:
                        dados = linha.strip().split(';')
                        cpf, nome, telefone, endereco = dados
                        self.__dados.insert(self.__dados.size() + 1, Usuario(cpf, nome, telefone, endereco))
        except FileNotFoundError:
            arquivo = open('cadastros.csv', 'w')
            arquivo.write('CPF;Nome;Telefone;Endere√ßo\n')
        except Exception as e:
            print(f"Erro ao ler o arquivo: {e}")


    def __str__(self) -> str:
        return self.__dados.__str__()

    def print(self) -> None:
        print(self.__str__())


def dataframe():
    caminho = '/content/cadastros.csv'
    df = pd.read_csv(caminho, sep=';')
    display(df)

def clear_terminal():
    os.system('clear')

def menu (cadastro):
    print("\n+-------MENU------+")
    print("|1-Inserir usu√°rio|")
    print("|2-Remover usu√°rio|")
    print("|3-Atualizar dados|")
    print("|4-Consulta  √∫nica|")
    print("|5-Listar Usu√°rios|")
    print("+-----------------+\n")

    select = 0
    select = int(input('Digite sua op√ß√£o: '))

    while select < 1 or select > 5:
        select = int(input('Op√ß√£o inv√°lida, digite novamente sua op√ß√£o: '))

    print('\n\n')

    if select == 1:
        cpf = input("Digite o CPF do usu√°rio: ")
        nome = input("Digite o nome do usu√°rio: ")
        telefone = input("Digite o telefone do usu√°rio: ")
        endereco = input("Digite o endere√ßo do usu√°rio: ")
        cadastro.adicionarUsuario(cpf, nome, telefone, endereco)
        cadastro.salvarArquivo()

    elif select == 2:
        cpf = input("Digite o CPF do usu√°rio a ser removido: ")
        cadastro.removerUsuario(cpf)
        cadastro.salvarArquivo()

    elif select == 3:
        cpf = input("Digite o CPF do usu√°rio a ser atualizado: ")
        nome = input("Digite o novo nome do usu√°rio: ")
        telefone = input("Digite o novo telefone do usu√°rio: ")
        endereco = input("Digite o novo endere√ßo do usu√°rio: ")
        cadastro.atualizarUsuario(cpf, nome, telefone, endereco)
        cadastro.salvarArquivo()

    elif select == 4:
        cpf = input("Digite o CPF do usu√°rio a ser consultado: ")
        usuario = cadastro.consultarUsuario(cpf)
        if (usuario != None):
            print(f'\nCPF: {usuario.getCpf()}, Nome: {usuario.getNome()}, Telefone: {usuario.getTelefone()}, Endere√ßo: {usuario.getEndereco()} \n')
        else:
            print('\nN√£o existe nenhum usu√°rio cadastrado com este CPF\n')

    elif select == 5:
        dataframe()

    print('\n\n')


if __name__ == "__main__":
    # Exemplo de inser√ß√£o de cadastros
    """
    a = Cadastro()
    a.lerArquivo()
    a.adicionarUsuario("12345678901", "Jo√£o Silva", "123-456-7890", "Rua A, 123")
    a.adicionarUsuario("98765432101", "Maria Santos", "987-654-3210", "Avenida X, 456")
    a.adicionarUsuario("55555555555", "Lucas Pereira", "555-555-5555", "Rua B, 789")
    a.adicionarUsuario("11112222333", "Ana Oliveira", "111-222-3333", "Rua C, 987")
    a.adicionarUsuario("99998888777", "Pedro Mendes", "999-888-8777", "Rua D, 654")
    a.salvarArquivo()

    """


````
